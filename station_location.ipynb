```python
df1=pd.read_csv(r"D:\pr2\Data warehouse\Sorgente\201901-citibike-tripdata.csv")
df2=pd.read_csv(r"D:\pr2\Data warehouse\Sorgente\201902-citibike-tripdata.csv")
df3=pd.read_csv(r"D:\pr2\Data warehouse\Sorgente\201903-citibike-tripdata.csv")
df4=pd.read_csv(r"D:\pr2\Data warehouse\Sorgente\201904-citibike-tripdata.csv")
df5=pd.read_csv(r"D:\pr2\Data warehouse\Sorgente\201905-citibike-tripdata.csv")
df6=pd.read_csv(r"D:\pr2\Data warehouse\Sorgente\201906-citibike-tripdata.csv")
df7=pd.read_csv(r"D:\pr2\Data warehouse\Sorgente\201907-citibike-tripdata.csv")
df8=pd.read_csv(r"D:\pr2\Data warehouse\Sorgente\201908-citibike-tripdata.csv")
df9=pd.read_csv(r"D:\pr2\Data warehouse\Sorgente\201909-citibike-tripdata.csv")
df10=pd.read_csv(r"D:\pr2\Data warehouse\Sorgente\201910-citibike-tripdata.csv")
df11=pd.read_csv(r"D:\pr2\Data warehouse\Sorgente\201911-citibike-tripdata.csv")
df12=pd.read_csv(r"D:\pr2\Data warehouse\Sorgente\201912-citibike-tripdata.csv")
```

```python
df = pd.concat([df1, df2, df3, df4, df5, df6,df7,df8,df9,df10,df11,df12])
len(df)
```

```python
df['age'] = 2019 - df['birth year']
df.head()
```

```python
df.dtypes
```

```python
df['category'] = ["0-25" if x<25 else "25-34" if 25<=x<=34 else "35-44" if 35<=x<=44 else "45-54" if 45<=x<=54 else "55-64" if 55<=x<=64 else "65-74" if 65<=x<=74 else "75+" for x in df['age']]
df['gender'] = ["Male" if x==1 else "Female" if x==2 else "Unknown" for x in df['gender']]
df.head()
```

```python
df.rename(columns={"tripduration": "trip duration [s]"})
df['starttimemonth'] = pd.DatetimeIndex(df['starttime']).month
df['stoptimemonth'] = pd.DatetimeIndex(df['stoptime']).month
```

```python
df_sc = df.drop(columns=['start station id','end station id','tripduration', 'starttime', 'stoptime', 'bikeid','usertype','birth year'])
df_sc.head()
```

```python
df_start = df_sc.drop(columns=['end station name','end station latitude','end station longitude','stoptimemonth'])
df_start.head()
```

```python
df_end = df_sc.drop(columns=['start station name','start station latitude','start station longitude','starttimemonth'])
df_end.head()
```

```python
df_start.rename(columns={"start station name": "station name",'start station latitude': 'station latitude','start station longitude':'station longitude','starttimemonth': 'month'}, inplace=True)
df_start.head()
```

```python
df_end.rename(columns={"end station name": "station name",'end station latitude': 'station latitude','end station longitude':'station longitude', 'stoptimemonth': 'month'}, inplace=True)
df_end.head()
```

Creazione station_location.csv

```python
start_station_location_sc= df_start[['station name', 'station latitude','station longitude']].copy()
end_station_location_sc= df_end[['station name', 'station latitude','station longitude']].copy()
station_location_sc = pd.concat([start_station_location_sc,end_station_location_sc])
```

```python
station_location = station_location_sc.drop_duplicates(subset="station name")
len(station_location)
```

```python
station_location.to_csv(r"D:\pr2\Data warehouse\station_location.csv", index=False)
```
